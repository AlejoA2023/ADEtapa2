```python
# Carga de Librerias

import matplotlib.pyplot as plt
from sklearn import linear_model
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("C:/Datos1/archive/data.csv", sep=",")
```

```python
data
```

```python
reglineal = linear_model.LinearRegression()
```

```python
# Agregar los datos a un array

x = data["metro"].values.reshape((-1,1))
y = data["precio"]
```

```python
# Crear el modelo

modelo = reglineal.fit(x,y)

print("Interseccion (b)", modelo.intercept_)

# Imprimir la pendiente

print("Pendiente (m)", modelo.coef_)
```

```python
# Predicciones

pred = modelo.predict(x)
pred
```

```python
# Graficar

data.plot.scatter(x="metro", y="precio", color="green", label='Datos Originales')
plt.scatter(x,pred, color="blue")
plt.plot(x,pred, color="red", linewidth=2, label='Linea de Regresion')
plt.legend()
plt.title("Ejercicio Regresión Lineal Simple")
plt.show
```
